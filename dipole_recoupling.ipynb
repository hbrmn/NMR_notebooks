```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as ss
```

## Bessel function

$$ \frac{\Delta S}{S_{0}} (\text{random}) = \frac{1}{2S+1}  \Bigg[ 2S  - \frac{\pi}{4(2S+1)} \sum \limits _{k=1} ^{2S} \big[4S-2(k-1)\big] J_{1/4} \big(k \sqrt{2} \lambda \big) J_{-1/4} \big(k \sqrt{2} \lambda \big) \Bigg]$$

### Representation of individual +0.25 and -0.25 Bessel functions

```python
def bessel_p25(xaxis):
    return ss.jv(0.25,xaxis)
def bessel_m25(xaxis):
    return ss.jv(-0.25,xaxis)
def bessel_comb(xaxis):
    return  bessel_p25(xaxis)+bessel_m25(xaxis)
xaxis = np.linspace(1,40,10000)
plt.figure(1)
plt.plot(xaxis, -bessel_p25(xaxis), color='r') 
plt.plot(xaxis, -bessel_m25(xaxis), color='b')
plt.rc('lines', linewidth=5)
#plt.plot(xaxis, -bessel_comb(xaxis), color='k')
#plt.savefig("Bessel_demo.png",format='png', dpi=600, bbox_inches='tight')
```

### Representation of random dipolar dephasing Bessel function [A]

```python
def bessel_3half(xaxis, dip_const):
    return (0.25*
                (3 -(np.pi*np.sqrt(2))/16 *
                 ((6*ss.jv(0.25, 1*np.sqrt(2)*dip_const*xaxis)
                   *ss.jv(-0.25, 1*np.sqrt(2)*dip_const*xaxis))
                  + (4*ss.jv(0.25, 2*np.sqrt(2)*dip_const*xaxis)
                     *ss.jv(-0.25, 2*np.sqrt(2)*dip_const*xaxis))
                  + (2*ss.jv(0.25, 3*np.sqrt(2)*dip_const*xaxis)
                     *ss.jv(-0.25, 3*np.sqrt(2)*dip_const*xaxis)))))
def bessel_5half(xaxis, dip_const):
    return ((1/6)*
                (5 -(np.pi*np.sqrt(2))/24 *
                 ((10*ss.jv(0.25, 1*np.sqrt(2)*dip_const*xaxis)
                   *ss.jv(-0.25, 1*np.sqrt(2)*dip_const*xaxis))
                  + (8*ss.jv(0.25, 2*np.sqrt(2)*dip_const*xaxis)
                     *ss.jv(-0.25, 2*np.sqrt(2)*dip_const*xaxis))
                  + (6*ss.jv(0.25, 3*np.sqrt(2)*dip_const*xaxis)
                     *ss.jv(-0.25, 3*np.sqrt(2)*dip_const*xaxis))
                  + (4*ss.jv(0.25, 4*np.sqrt(2)*dip_const*xaxis)
                     *ss.jv(-0.25, 4*np.sqrt(2)*dip_const*xaxis))
                  + (2*ss.jv(0.25, 5*np.sqrt(2)*dip_const*xaxis)
                     *ss.jv(-0.25, 5*np.sqrt(2)*dip_const*xaxis)))))

xaxis = np.linspace(0,0.008,100)
plt.figure(1)
plt.plot(xaxis*1000, bessel_3half(xaxis,500), color='k')
plt.plot(xaxis*1000, bessel_3half(xaxis,1500), color='g')
plt.plot(xaxis*1000, bessel_5half(xaxis,500), color='b') 
plt.xlim(0, 8)
plt.ylim(-0.05, 1)
plt.xlabel(r"NT$_{\mathrm{r}}$ / ms")
plt.ylabel(r"$\Delta \mathrm{S} / \mathrm{S}_{0}$")
plt.rc('lines', linewidth=2)
plt.savefig("Bessel_demo2.png",format='png', dpi=600, bbox_inches='tight')
```
